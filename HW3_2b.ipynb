```
"""
HW3_2b: Concentration of Pollutant
@author: Kiran Pandey
"""
#==============================================================================
#                   Modules
#==============================================================================
import numpy as np
import matplotlib.pyplot as plt

#==============================================================================
#                  parameters
#==============================================================================
T    = np.arange(0,  10000, 1)
V    = (1e+2) + (9.7680*T) 
P    = 4*T
Cp   = P/V

#==============================================================================
#                   Plot and save Solution
#==============================================================================
dir_out='result'
file_out1='Conc_Pollutant.png'

plt.figure(1,figsize=(8,8))
plt.title('Concentration of pollutant in the lake as a function of time ', fontsize=16)
plt.plot(T,Cp,'bo',ms=1)
plt.xlabel('Time, years')
plt.ylabel('Concentration, kg/m3')
plt.savefig('%s/%s'%(dir_out,file_out1))
```
