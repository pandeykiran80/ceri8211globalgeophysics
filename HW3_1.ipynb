```
# Pyth0n 3.7
"""
HW3_1: Evolution of Population of Rabbit and Fox
@author: Kiran Pandey
"""
#==============================================================================
#                   Modules
#==============================================================================
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
#==============================================================================
#                  parameters
#==============================================================================
R0 = 100
F0 = 100
Br = 0.01
Bf = 0.005
dr = 0.005
df = 0.01
Er = 0.0001
Ef = 0.0001
t0 = 0
#==============================================================================
#                  Computation
#==============================================================================
P0 = [R0, F0]  
def fct_dxdt(y, t0):
    return np.array([ Br*y[0] - dr*y[0] - Ef*y[0]*y[1] ,
                  Bf*y[1] - df*y[1] + Er*y[0]*y[1] ])
  
t = np.arange(0,  5000, 0.5)     # time, days  
           
y = odeint(fct_dxdt,P0,t)    
rabbit, fox = y.T
#==============================================================================
#                  Plot
#==============================================================================
dir_out='result'
file_out1='RabbitFoxPopn.png'

plt.figure(1,figsize=(8,8))
plt.title(' Evolution of Rabbit and Fox Population ', fontsize=16)
plt.plot(t,fox,'k-',ms=2)
plt.plot(t,rabbit,'b-',ms=2)
plt.xlabel('time, days')
plt.ylabel(' Population ')
plt.show()
#plt.xlim(0,5)
#plt.ylim(0.705,0.715)
plt.savefig('%s/%s'%(dir_out,file_out1))
```
