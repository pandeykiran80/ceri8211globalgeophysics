```
# Pyth0n 3.7
"""
HW3_5a: Sr and Rb isotopic compositions (age of intrusion)
@author: Kiran Pandey
"""
#==============================================================================
#                   Modules
#==============================================================================
import numpy as np
import matplotlib.pyplot as plt

#==============================================================================
#                  parameters
#==============================================================================
Sr87_Sr86 = [0.7117, 0.7095, 0.7092, 0.7083, 0.7083, 0.7082]
Rb87_Sr86 = [3.65, 1.80, 1.48, 0.82, 0.66, 0.74]

#==============================================================================
#                  Computation
#==============================================================================
isochron = np.polyfit(Rb87_Sr86,Sr87_Sr86,1 )
x_axis = np.arange(0,  5, .5)
y_axis = np.polyval(isochron, x_axis)
slope, intercept = np.polyfit(Rb87_Sr86,Sr87_Sr86,1 )
print('Slope of the line is', slope)
print('Intercept of the line is', intercept)

#==============================================================================
#                  Plot
#==============================================================================
dir_out='result'
file_out1='Rb_SrIsochron.png'

plt.figure(1,figsize=(8,8))
plt.title(' Rb_Sr isochron diagram ', fontsize=16)
plt.plot(x_axis,y_axis,'k-',ms=2)
plt.plot(Rb87_Sr86,Sr87_Sr86,'b*',ms=10)
plt.xlabel('Rb87/Sr86')
plt.ylabel(' Sr87/Sr86 ')
plt.xlim(0,5)
plt.ylim(0.705,0.715)
plt.savefig('%s/%s'%(dir_out,file_out1))
```
