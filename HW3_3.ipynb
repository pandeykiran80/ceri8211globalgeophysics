```
# Pyth0n 3.7
"""
HW3_3a: U-Pb Concordia Discordia
@author: Kiran Pandey
"""
#==============================================================================
#                   Modules
#==============================================================================
import numpy as np
import matplotlib.pyplot as plt

#==============================================================================
#                  parameters
#==============================================================================
Pb206_U238 = (10**6)*np.array([4233, 4201, 3780, 1713, 2595, 1177, 4356, 4235, 3675, 4126, 2604, 1961, 4364, 3355])
Pb207_U235 = (10**6)*np.array([4321, 4303, 4153, 3155, 3672, 2808, 4355, 4307, 4078, 4287, 3651, 3317, 4392, 3985])
lambda_U235 = 9.8485 * (10**-10)
lambda_U238 = 1.5513 * (10**-10)
t = np.arange(0,  5000, 100)
T = t*(10**6)

a_U235 = np.exp(lambda_U235*T) - 1  # U235
b_U238 = np.exp(lambda_U238*T) - 1  # U238

x_U235 = np.exp(lambda_U235*Pb207_U235) -1
y_U238 = np.exp(lambda_U238*Pb206_U238) -1

p = np.polyfit(x_U235,y_U238,1)
line = np.poly1d(p)
j = np.linspace(-5, 100, 104)

#==============================================================================
#                   Plot and save Solution
#==============================================================================
#dir_out='result'
#file_out1='U_Pb_CD.png'

plt.figure(1,figsize=(8,8))
plt.title('Concordia-Discordia Plor: U-Pb ', fontsize=16)
line1=plt.plot(a_U235, b_U238, 'b-')
line2=plt.plot( j, line(j),  'k-')
line3=plt.plot(x_U235, y_U238, 'r*')
plt.xlabel('Pb207/U235')
plt.ylabel('Pb206/U238')
plt.legend((line1, line2, line3), ('Concordia', 'Dicordia', 'Zircon'))
#plt.legend('Concordia','Dicordia','Zircon')
plt.show()
#plt.xlim(0,5)
#plt.ylim(0.705,0.715)
#plt.savefig('%s/%s'%(dir_out,file_out1))



```
